```python
# Экспериментальные данные
exp_data = {
    4.78: 4, 4.79: 1, 4.8: 0, 4.81: 4, 4.82: 1, 4.83: 0, 4.84: 4,
    4.85: 3, 4.86: 0, 4.87: 4, 4.88: 6, 4.89: 0, 4.9: 2, 4.91: 7, 
    4.92: 0, 4.93: 2, 4.94: 5, 4.95: 0, 4.96: 3, 4.97: 8, 4.98: 0, 
    4.99: 0, 5.0: 12, 5.01: 0, 5.02: 0, 5.03: 12, 5.04: 0, 5.05: 0, 
    5.06: 6, 5.07: 2, 5.08: 0, 5.09: 3, 5.1: 1, 5.11: 0, 5.12: 2, 
    5.13: 1, 5.14: 0, 5.15: 2, 5.16: 3, 5.17: 0, 5.18: 0, 5.19: 0, 
    5.2: 0, 5.21: 0, 5.22: 1, 5.23: 0, 5.24: 0, 5.25: 1
}
```

```python
# Разобраться с округлением!!!
data = []
for k, v in exp_data.items():
    data.extend([k]*v)
N = len(data)
print('Количество измерений N =', N)
print('Сумма =', sum(data))
T = round(sum(data)/N, 2)
print('Среднее T =', T)
table = ''
for i, t in enumerate(data):
    table += f'{i + 1} & {t:.2f} & {(t - T):.2f} & {round((t - T)**2, 4):.4f} \\\\ \\hline '
sum_abs_fallibility = round(sum(t - T for t in data), 2)
print('Сумма погрешностей в подвал =', sum_abs_fallibility)
sum_abs_square_fallibility = sum(round((t - T)**2, 4) for t in data)
print('Сумма погрешностей в квадрате =', sum_abs_square_fallibility)
print(table)
```
