```python
# Экспериментальные данные
exp_data = {
    4.78: 4, 4.79: 1, 4.8: 0, 4.81: 4, 4.82: 1, 4.83: 0, 4.84: 4,
    4.85: 3, 4.86: 0, 4.87: 4, 4.88: 6, 4.89: 0, 4.9: 2, 4.91: 7, 
    4.92: 0, 4.93: 2, 4.94: 5, 4.95: 0, 4.96: 3, 4.97: 8, 4.98: 0, 
    4.99: 0, 5.0: 12, 5.01: 0, 5.02: 0, 5.03: 12, 5.04: 0, 5.05: 0, 
    5.06: 6, 5.07: 2, 5.08: 0, 5.09: 3, 5.1: 1, 5.11: 0, 5.12: 2, 
    5.13: 1, 5.14: 0, 5.15: 2, 5.16: 3, 5.17: 0, 5.18: 0, 5.19: 0, 
    5.2: 0, 5.21: 0, 5.22: 1, 5.23: 0, 5.24: 0, 5.25: 1
}
```

```python
# Разобраться с округлением!!!
data = []
for k, v in exp_data.items():
    data.extend([k]*v)
N = len(data)
print('Количество измерений N =', N)
print('Сумма =', sum(data))
T = round(sum(data)/N, 4)
print('Среднее t =', T)
table = ''
for i, t in enumerate(data):
    table += f'{i + 1} & {t:.2f} & {(t - T):.2f} & {round((t - T)**2, 4):.4f} \\\\ \\hline '
    
sum_abs_fallibility = round(sum(t - T for t in data), 2)
print('Сумма погрешностей в подвал =', sum_abs_fallibility)

sum_abs_square_fallibility = round(sum(((t - T)**2) for t in data), 4)
print('Сумма погрешностей в квадрате =', sum_abs_square_fallibility)

sigma_N = round((1/99 * 1.1114)**0.5, 4)
print(f'\\sigma_N = {sigma_N:.4f} c')

rho_max = round(1/(sqrt(2*pi) * sigma_N), 4)
print(f'\\rho_{'{max}'} = {rho_max:.4f} c^{'{-1}'}')
print(table)
```

```python
t_min, t_max = min(data), max(data)
print('t_{min} =', t_min, '\nt_{max} =', t_max)
print('sqrt_N =', sqrt(N))
# Подобрать m
m = 10
print((t_max - t_min) / (m - 1))
m_t_delta = 0.06
m_t_min = 4.74
print([round(m_t_min + m_t_delta * i, 2)  for i in range(0, m, 1)])
```

```python
table2 = ''
gist_data = []
for i in range(0, m - 1):
    start = round(m_t_min + m_t_delta * i, 2)
    finish = round(m_t_min + m_t_delta * (i + 1), 2)
    DeltaN = sum(1 for x in data if start <= x < finish)
    frequency = round(DeltaN / (N * m_t_delta), 2)
    ttt = round(start + m_t_delta / 2, 2)
    stolb_rho = (1 / sqrt(2*pi) / sigma_N) * exp(-(ttt - T)**2/(2*sigma_N**2))
    gist_data.append((ttt, frequency, stolb_rho))
    dan = (f'{start:.2f}', f'{finish:.2f}', f'{DeltaN}', 
           f'{frequency:.2f}', f'{ttt:.2f}', f'{stolb_rho:.2f}')
    table2 += (str(dan[0]) +' & \\multirow{2}{*}{'+ str(dan[2]) +'} & \\multirow{2}{*}{'+ 
              str(dan[3]) +'} & \\multirow{2}{*}{'+ str(dan[4]) +'} & \\multirow{2}{*}{'+ 
              str(dan[5]) +'} \\\\ \\cline{1-1}'+ str(dan[1]) +' & & & & \\\\ \\hline ')
print(table2)
    
```

```python
dN1 = sum(1 for x in data if T - sigma_N < x < T + sigma_N) 
dN2 = sum(1 for x in data if T - 2*sigma_N < x < T + 2*sigma_N) 
dN3 = sum(1 for x in data if T - 3*sigma_N < x < T + 3*sigma_N) 

table3 = f'''${'{\\langle t \\rangle}'}_N \\pm \\sigma_N$ & {T - sigma_N:.4f} & {T + sigma_N:.4f} & {dN1} & {dN1 / N:.2f} & 0.683 \\\\ \\hline
${'{\\langle t \\rangle}'}_N \\pm 2\\sigma_N$ & {T - 2*sigma_N:.4f} & {T + 2*sigma_N:.4f} & {dN2} & {dN2 / N:.2f} & 0.954 \\\\ \\hline
${'{\\langle t \\rangle}'}_N \\pm 3\\sigma_N$ & {T - 3*sigma_N:.4f} & {T + 3*sigma_N:.4f} & {dN3} & {dN3 / N:.2f} & 0.997 \\\\ \\hline'''
print(table3)
```

```python
sigma_pog = round(sqrt(sum_abs_square_fallibility / N / (N + 1)), 4)
print('Cреднеквадратичное отклонение среднего значения =', sigma_pog)
student = 1.9840
t_pog = round(sigma_pog * student, 2)
print('Абсолютная погрешность =', t_pog)
E_pog = round(t_pog / T * 100, 2)
print('Относительная погрешность =', E_pog)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import make_interp_spline

# Создаем график
plt.figure(figsize=(10, 6))

# Извлекаем границы и высоты
left_edges = [x[0] for x in gist_data]
heights = [x[1] for x in gist_data]
widths = [m_t_delta for x in gist_data]
# Создаем гистограмму
plt.bar(left_edges, heights, width=widths, 
        color='skyblue', edgecolor='black', label='Гистограмма')

# Ваши данные
x = np.array([point[0] for point in gist_data])
y = np.array([point[2] for point in gist_data])
# Создаем сглаженную кривую с правильными границами
spline = make_interp_spline(x, y, k=3, bc_type='not-a-knot')
x_new = np.linspace(min(x), max(x), 300)
y_new = spline(x_new)
# Создаем график
plt.plot(x_new, y_new, 'b-', label='Кривая по Гаусу')
plt.scatter(x, y, color='red', label='Исходные точки по Гаусу')


# Добавляем сетку
plt.grid(True, which='both', linestyle='--', alpha=0.7)
# Настраиваем подписи осей
plt.xlabel('t, c', x=1.01, ha='right', va='bottom')
plt.ylabel('ΔN/NΔt, Гц', y=1.03, ha='left', va='top', rotation=0)
# Устанавливаем масштаб по оси X
plt.gca().xaxis.set_major_locator(plt.MultipleLocator(m_t_delta / 2))
# Добавляем заголовок
plt.title('График распределения')
# plt.legend()

# Показываем график
plt.show()
```
