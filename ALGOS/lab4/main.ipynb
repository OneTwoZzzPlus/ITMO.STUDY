```python
from random import randint

N = 20 # int(input())
s = [randint(-100, 100) for i in range(N)]


def seq(l):
    maxim = 0
    p = [0, 0]
    n = 1
    k, m = 0, 0
    for i in range(1, len(l)):
        if l[i] >= l[i - 1]:
            n += 1
            m += 1
        else:
            if maxim < n:
                maxim = n
                p = [k, m]
            n = 1
            k = i
            m = i
    return l[p[0]:p[1] + 1]


print(s)
print(seq(s))
```

### Генерация строк

```python
def gen_primes(n: int) -> str:
    D, q, primes = {}, 2, ''
    while n > 0:
        if q not in D:
            n -= 1
            primes += str(q)     
            D[q * q] = [q]
        else:
            for p in D[q]:
                D.setdefault(p + q, []).append(p)
            del D[q]
        q += 1
    return primes

def gen_fibonachi(n: int) -> str:
    fib1, fib2, fibonachi = 1, 1, '11'
    for _ in range(n-2):
        fib1, fib2 = fib2, fib1 + fib2
        fibonachi += str(fib2)
    return fibonachi

fibonachi = gen_fibonachi(500)
primes = gen_primes(500)

print('fibonachi', fibonachi)
print('prime', primes)
```

### Решатель через find

```python
def solver(s: str) -> list:
    ans = []
    for i in range(10, 100):
        template = str(i)
        k = 0
        index = -1
        while True:
            index = s.find(template, index + 1)
            if index == -1:
                break
            k += 1
        ans.append((str(template), k))

    return sorted(ans, key=lambda x: -x[1])

fibonachi_answer = solver(fibonachi)
primes_answer = solver(primes)

print('fibonachi', *fibonachi_answer)
print('prime', *primes_answer)
```

### Тестер

```python
timing = {}

def test(func, text: str, answer: str, display_err=10, display_ok=3):
    global timing
    from time import time
    timer = time()
    
    ans = []
    for i in range(10, 100):
        pattern = str(i)
        ans.append((pattern, len(func(text, pattern))))
    
    res_time = time() - timer
    ans.sort(key=lambda x: -x[1])
    if ans == answer:
        print(func.__name__, "correct", f"{res_time:.5f}s", ans[:display_ok])
        
        timing.setdefault(func.__name__, {})[len(text)] = f"{res_time:.5f}s"
    else:
        print(func.__name__, "ERROR", ans[:display_err])
        print('ANSWER', answer[:display_err])
```

### Поиск максимального элемента

```python
def search(func, text: str, count=3):
    mx, m = [0] * count, [''] * count
    for i in range(10, 100):
        pattern = str(i)
        result = len(func(text, pattern))
        
        for k in range(count):
            if result > mx[k]:
                mx.insert(k, result)
                m.insert(k, pattern)
                break
                
    for k in range(count):
        print(f'\tШаблон {m[k]} встречается {mx[k]} раз')
```

### Алгоритм наивный

```python
def naive(text: str, pattern: str) -> list[int]:
    result = []
    for i in range(len(text)-1):
        if pattern == text[i:i+2]:
            result.append(i)
    return result
         

print("Проверка корректности:")
test(naive, primes, primes_answer)
test(naive, fibonachi, fibonachi_answer)
print("Простые числа:")
search(naive, primes)
print("Числа Фиббоначи:")
search(naive, fibonachi)
```

### Алгоритм Рабина-Карпа
Для текста из цифр: размер алфавита x = 10, размер шаблона m = 2 или 1, индекс цифры - цифра =)

```python
def hash_2digit(s: str) -> int:
    return int(s) if s[0] != s[1] else int(s[0])
    
    
def hash_digit(s: str) -> int:
    return int(s)

    
def RabinKarp(text: str, pattern: str, whash=hash_digit) -> list[int]:
    result = []
    pattern_hash = whash(pattern)
    for i in range(len(text)-1):
        if pattern_hash == whash(text[i:i+2]):
            if pattern == text[i:i+2]:
                result.append(i)
    return result


print("Проверка корректности:")
test(RabinKarp, primes, primes_answer)
test(RabinKarp, fibonachi, fibonachi_answer)
print("Простые числа:")
search(RabinKarp, primes)
print("Числа Фиббоначи:")
search(RabinKarp, fibonachi)
```

### Алгоритм Бойера-Мура (эвристика плохого символа)

```python
def BoyerMurBadChar(text: str, pattern: str) -> list[int]:
    indexes = {pattern[i]: i for i in range(len(pattern))}
    result = []
    shift = 0

    while shift <= len(text) - len(pattern):
        current = len(pattern) - 1
        
        while current >= 0 and text[shift + current] == pattern[current]:
            current -= 1
            
        if current == -1:
            result.append(shift)
            shift += (
                len(pattern) - indexes.get(text[shift + len(pattern)], -1) 
                if shift + len(pattern) < len(text) else 1
            )
        else:
            err_char = text[shift + current]
            shift += max(1, current - indexes.get(err_char, -1))
            
    return result


print("Проверка корректности:")
test(BoyerMurBadChar, primes, primes_answer)
test(BoyerMurBadChar, fibonachi, fibonachi_answer)
print("Простые числа:")
search(BoyerMurBadChar, primes)
print("Числа Фиббоначи:")
search(BoyerMurBadChar, fibonachi)
```

### Алгоритм Кнута — Морриса — Пратта

```python
def pr(st):
    sp = []
    for j in range(1, len(st)+1):
        sp.append(0)
        for i in range(len(st[:j])):
            if st[:j][:i] == st[:j][-1*i:]:
                sp[-1] = max(sp[-1], i)
    return sp


def KMP(text: str, pattern: str) -> list[int]:
    pref = pr(str(pattern))
    result = []
    i = 0
    j = 0
    while i < len(text):
        if text[i] == str(pattern)[j]:
            i += 1
            j += 1
            if j == 2:
                result.append(i)
                j = pref[j-1]
        else:
            if j > 0:
                j = pref[j-1]
            else:
                i += 1
    return result
                
print("Проверка корректности:")
test(KMP, primes, primes_answer)
test(KMP, fibonachi, fibonachi_answer)
print("Простые числа:")
search(KMP, primes)
print("Числа Фиббоначи:")
search(KMP, fibonachi)
```
