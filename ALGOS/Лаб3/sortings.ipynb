### Тестирующие функции

```python
import timeit

def generate_lst(a: int=-100, b: int=100, length: int=200):
    """Генерирует список случайных чисел"""
    return [__import__("random").randint(a, b) for _ in range(length)]


def test(func, lst=generate_lst(-10000, 10000, 1000)):
    # Тестируем корректность
    func_lst = func(lst.copy())
    sorted_lst = list(sorted(lst))
    status = "RIGHT" if (correct := (sorted_lst == func_lst)) else "ERROR"
    # Отображаем информацию
    print(f'Algorythm:\t{func.__name__}\nCorrectness:\t{status}')
    print('LIST', lst)          # Изначальный
    print('SORT', sorted_lst)   # Отсортирован .sort()
    print('FUNC', func_lst)     # Результат работы функции
    # Поиск ошибок
    if not correct:
        print('=== ERROR TRACER ===')
        if len(sorted_lst) != len(func_lst):
            print(
f'Разная длина списков: input={len(sorted_lst)}, output={len(func_lst)}'
            )
        elif lst == func_lst:
            print(f'Вернулся изначальный список!')
        elif list(reversed(sorted_lst)) == func_lst:
            print(f'Список отсортирован в другую сторону!')
        else:
            print('Ошибки индекс=верное=полученное:')
            for i in range(len(sorted_lst)):
                if sorted_lst[i] != func_lst[i]:
                    print(i, sorted_lst[i], func_lst[i], sep='=', end=', ')
            print()


def time_test(funcs, repeat=100, length: int=200, a: int=-10000, b: int=10000):
    print(f'PROPERTIES Repeat: {repeat}, Length: {length}')

    for func in funcs:

        # timeit.repeat
        time = timeit.timeit(
            stmt=f'{func.__name__}(generate_lst({a}, {b}, {length}))',
            setup=f'from __main__ import generate_lst, {func.__name__}',
            number=repeat
        )

        avg_exec_time = round(time/repeat, 8)
        print(f'TEST:\t{func.__name__}\t{avg_exec_time}s')
    print()
```

### Быстрая сортировка
O(n\*logn) best(n\*logn) worst(n^2)

```python
def QuickSort(a):
    if len(a) < 2:
        return a
    mid = len(a) // 2
    less, equals, more = [], [], []
    for x in a:
        if x == a[mid]:
            equals.append(x)
        elif x > a[mid]:
            more.append(x)
        else:
            less.append(x)
    return QuickSort(less) + equals + QuickSort(more)

test(QuickSort)
```

### Сортировка расчёсткой
O(𝑛^2/2^𝑝) best(n*logn) worst(n^2)

```python
def CombSort(a):
    length, step = len(a), len(a)
    while step > 1 or flag:
        if step > 1:
            step = int(step / 1.25)
        flag = False
        for i in range(0, length - step, step):
            if a[i] > a[i + step]:
                a[i], a[i + step] = a[i + step], a[i]
                flag = True
    return a

test(CombSort)
```

### Тесты через timeit: QuickSort и CombSort

```python
sortings = [QuickSort, CombSort]

time_test(sortings, repeat=1000, length=10)
time_test(sortings, repeat=1000, length=100)
time_test(sortings, repeat=100, length=1000)
# time_test(sortings, repeat=2, length=10000)
# time_test(sortings, repeat=1, length=20000)
```

### Блочная сортировка

```python
def BlockSortV2(sp, mn=-10000, mx=10000, mtc=100):
    mt=[[] for _ in range(mtc)]
    for i in sp:
        mt[int(len(mt)*(i-mn)//(mx-mn+1))].append(i)
    for i in range(len(mt)):
        if len(mt) >= ((len(sp)**0.5) * 2):
            mt.append(BlockSortV2(mt[i]))
            del mt[i]
    for i in range(len(mt)):
        mt[i] = sorted(mt[i])
    rt = []
    for i in mt:
        rt += i
    return rt


test(BlockSortV2)
```

### Пирамидальная сортировка

```python
def heapify(sp, n, i):
    mx = i
    l = 2 * i + 1
    r = 2 * i + 2
    if l < n and sp[i] < sp[l]:
        mx = l
    if r < n and sp[mx] < sp[r]:
        mx = r
    if mx != i:
        (sp[i], sp[mx]) = (sp[mx], sp[i])
        heapify(sp, n, mx)


def HeapSort(sp):
    n = len(sp)
    for i in range(n // 2, -1, -1):
        heapify(sp, n, i)
    for i in range(n - 1, 0, -1):
        sp[i], sp[0] = sp[0], sp[i]
        heapify(sp, i, 0)
    return sp

test(HeapSort)
```

### Сортировка слиянием

```python
def MergeSort(sp):
    if len(sp) <= 1:
        return sp
    left = sp[:len(sp) // 2]
    right = sp[len(sp) // 2:]
    return merge(MergeSort(left), MergeSort(right))


def merge(l, r):
    res = []
    i = 0
    j = 0
    while i < len(l) and j < len(r):
        if l[i] < r[j]:
            res.append(l[i])
            i += 1
        else:
            res.append(r[j])
            j += 1
    res += l[i:] + r[j:]
    return res


test(MergeSort)
```
